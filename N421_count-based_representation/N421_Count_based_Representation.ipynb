> _â€œíŠ¹ì„±ì˜ ê°œìˆ˜ê°€ ì„ í˜•ì ìœ¼ë¡œ ëŠ˜ì–´ë‚  ë•Œ ë™ì¼í•œ ì„¤ëª…ë ¥ì„ ê°€ì§€ê¸° ìœ„í•´ í•„ìš”í•œ ì¸ìŠ¤í„´ìŠ¤ì˜ ìˆ˜ëŠ” ì§€ìˆ˜ì ìœ¼ë¡œ ì¦ê°€í•œë‹¤.<br/>

```python
import spacy
from spacy.tokenizer import Tokenizer

nlp = spacy.load("en_core_web_sm")

tokenizer = Tokenizer(nlp.vocab)
```

```python
sent1 = "I am a student."
sent2 = "J is the alphabet that follows i."
sent3 = "Is she a student trying to become a data scientist?"

sent_lst = [sent1, sent2, sent3]

total_tokens = []

for i, sent in enumerate(tokenizer.pipe(sent_lst)):
    sent_token = [token.text for token in sent]
    total_tokens.extend(sent_token)
    print(f"sent {i} : {sent_token}")

set(total_tokens)
print(f""" 
total_tokens : {set(total_tokens)}
#_of_total_tokens : {len(set(total_tokens))}
""")
```

```python
def word2idx(sent):
    sent_token = sent.split()
    return [1 if word in sent else 0 for word in set(total_tokens)]


sent1_idx = word2idx(sent1)
sent2_idx = word2idx(sent2)
sent3_idx = word2idx(sent3)
```

```python
import pandas as pd

dtm = pd.DataFrame([sent1_idx, sent2_idx, sent3_idx], columns=set(total_tokens))
dtm
```

```python
import re

def lower_and_regex(sentence):
    """
    ëª¨ë“  ëŒ€ë¬¸ìë¥¼ ì†Œë¬¸ìë¡œ ë³€ê²½ í›„
    ì •ê·œì‹ì„ ì´ìš©í•˜ì—¬ ì•ŒíŒŒë²³ ì†Œë¬¸ì ì´ì™¸ì˜ êµ¬ë‘ì  ì‚­ì œ
    """
    sentence = sentence.lower()
    sentence = re.sub(r"[^a-z ]", "", sentence)
    return sentence

prep_sent_lst = [lower_and_regex(sent) for sent in sent_lst]
total_tokens_prep = []

for i, prep_sent in enumerate(tokenizer.pipe(prep_sent_lst)):
    sent_token_prep = [token.text for token in prep_sent]
    total_tokens_prep.extend(sent_token_prep)
    print(f"sent {i} : {sent_token_prep}")

print(f""" 
total_tokens : {total_tokens_prep}
num_of_total_tokens : {len(set(total_tokens_prep))}
""")
```

### 3) ì •ê·œí‘œí˜„ì‹(Regex)

êµ¬ë‘ì ì´ë‚˜ íŠ¹ìˆ˜ë¬¸ì ë“± í•„ìš”ì—†ëŠ” ë¬¸ìê°€ ë§ë­‰ì¹˜ ë‚´ì— ìˆì„ ê²½ìš° í† í°í™”ê°€ ì œëŒ€ë¡œ ì´ë£¨ì–´ì§€ì§€ ì•ŠìŠµë‹ˆë‹¤. <br/> ì´ë¥¼ ì œê±°í•˜ê¸° ìœ„í•´ì„œ ì •ê·œí‘œí˜„ì‹(Regular Expression, Regex)ì„ ì‚¬ìš©í•©ë‹ˆë‹¤. <br/> ì •ê·œ í‘œí˜„ì‹ì€ ë¬¸ìì—´ì„ ë‹¤ë£¨ê¸° ìœ„í•œ ë§¤ìš° ì¤‘ìš”í•˜ê³ ë„ ê°•ë ¥í•œ ë„êµ¬ì´ì§€ë§Œ ë³µì¡í•˜ê¸° ë•Œë¬¸ì— ì˜ ì“°ë ¤ë©´ ë°˜ë³µì ì¸ ì‹¤ìŠµì´ í•„ìš”í•©ë‹ˆë‹¤.

`a-z`(ì†Œë¬¸ì), `A-Z`(ëŒ€ë¬¸ì), `0-9`(ìˆ«ì)ë¥¼ `^` ì œì™¸í•œ ë‚˜ë¨¸ì§€ ë¬¸ìë¥¼ `regex` ì— í• ë‹¹í•œ í›„ `.sub` ë©”ì„œë“œë¥¼ í†µí•´ì„œ ê³µë°± ë¬¸ìì—´ `""` ë¡œ ì¹˜í™˜í•´ì¤ë‹ˆë‹¤.

```python
# íŒŒì´ì¬ ì •ê·œí‘œí˜„ì‹ íŒ¨í‚¤ì§€ ì´ë¦„ì€ re ì…ë‹ˆë‹¤.
import re

# ì •ê·œì‹
# []: [] ì‚¬ì´ ë¬¸ìë¥¼ ë§¤ì¹˜, ^: not
regex = r"[^a-zA-Z0-9 ]"

# ì •ê·œì‹ì„ ì ìš©í•  ìŠ¤íŠ¸ë§
test_str = ("(Natural Language Processing) is easy!, DS!\n")

# ì¹˜í™˜í•  ë¬¸ì
subst = ""

result = re.sub(regex, subst, test_str)
```

`.sub` ë©”ì„œë“œ ì™¸ì— ë‹¤ë¥¸ ë©”ì„œë“œì— ëŒ€í•´ì„œë„ ì•Œì•„ë´…ì‹œë‹¤.

ì •ê·œí‘œí˜„ì‹ì„ ìˆ˜í–‰í•œ í›„ ì†Œë¬¸ìë¡œ í†µì¼í•˜ê³  ê³µë°± ë¬¸ìë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë‚˜ëˆ„ëŠ” í•˜ë‚˜ì˜ í•¨ìˆ˜ `tokenize`ë¥¼ ë§Œë“¤ì–´ ë³´ê² ìŠµë‹ˆë‹¤.

```python
def tokenize(text):
    """text ë¬¸ìì—´ì„ ì˜ë¯¸ìˆëŠ” ë‹¨ì–´ ë‹¨ìœ„ë¡œ listì— ì €ì¥í•©ë‹ˆë‹¤.
    Args:
        text (str): í† í°í™” í•  ë¬¸ìì—´
    Returns:
        list: í† í°ì´ ì €ì¥ëœ ë¦¬ìŠ¤íŠ¸
    """
    # ì •ê·œì‹ ì ìš©
    tokens = re.sub(regex, subst, text)

    # ì†Œë¬¸ìë¡œ ì¹˜í™˜
    tokens = tokens.lower().split()
    
    return tokens
```

```python
from collections import Counter

# Counter ê°ì²´ëŠ” ë¦¬ìŠ¤íŠ¸ìš”ì†Œì˜ ê°’ê³¼ ìš”ì†Œì˜ ê°¯ìˆ˜ë¥¼ ì¹´ìš´íŠ¸ í•˜ì—¬ ì €ì¥í•˜ê³  ìˆìŠµë‹ˆë‹¤.
# ì¹´ìš´í„° ê°ì²´ëŠ” .update ë©”ì†Œë“œë¡œ ê³„ì† ì—…ë°ì´íŠ¸ ê°€ëŠ¥í•©ë‹ˆë‹¤.
word_counts = Counter()

# í† í°í™”ëœ ê° ë¦¬ë·° ë¦¬ìŠ¤íŠ¸ë¥¼ ì¹´ìš´í„° ê°ì²´ì— ì—…ë°ì´íŠ¸ í•©ë‹ˆë‹¤. 
df['tokens'].apply(lambda x: word_counts.update(x))

# ê°€ì¥ ë§ì´ ì¡´ì¬í•˜ëŠ” ë‹¨ì–´ ìˆœìœ¼ë¡œ 10ê°œë¥¼ ë‚˜ì—´í•©ë‹ˆë‹¤
word_counts.most_common(10)
```

ìœ„ ì½”ë“œë¥¼ ë³€í˜•í•˜ì—¬ ì½”í¼ìŠ¤ì˜ ì „ì²´ ì›Œë“œ ì¹´ìš´íŠ¸, ë­í¬ ë“± ì •ë³´ê°€ ë‹´ê¸´ ë°ì´í„°í”„ë ˆì„ì„ ë¦¬í„´í•˜ëŠ” í•¨ìˆ˜ë¥¼ ë§Œë“¤ì–´ ë´…ì‹œë‹¤.<br/>
ì´ í•¨ìˆ˜ëŠ” í† í°í™”ëœ ë¬¸ì„œë“¤ì´ ë“¤ì–´ìˆëŠ” ì½”í¼ìŠ¤ë¥¼ ì…ë ¥ìœ¼ë¡œ ë°›ìŠµë‹ˆë‹¤.

```python
def word_count(docs):
    """ í† í°í™”ëœ ë¬¸ì„œë“¤ì„ ì…ë ¥ë°›ì•„ í† í°ì„ ì¹´ìš´íŠ¸ í•˜ê³  ê´€ë ¨ëœ ì†ì„±ì„ ê°€ì§„ ë°ì´í„°í”„ë ˆì„ì„ ë¦¬í„´í•©ë‹ˆë‹¤.
    Args:
        docs (series or list): í† í°í™”ëœ ë¬¸ì„œê°€ ë“¤ì–´ìˆëŠ” list
    Returns:
        list: Dataframe
    """
    # ì „ì²´ ì½”í¼ìŠ¤ì—ì„œ ë‹¨ì–´ ë¹ˆë„ ì¹´ìš´íŠ¸
    word_counts = Counter()

    # ë‹¨ì–´ê°€ ì¡´ì¬í•˜ëŠ” ë¬¸ì„œì˜ ë¹ˆë„ ì¹´ìš´íŠ¸, ë‹¨ì–´ê°€ í•œ ë²ˆ ì´ìƒ ì¡´ì¬í•˜ë©´ +1
    word_in_docs = Counter()

    # ì „ì²´ ë¬¸ì„œì˜ ê°¯ìˆ˜
    total_docs = len(docs)

    for doc in docs:
        word_counts.update(doc)
        word_in_docs.update(set(doc))

    temp = zip(word_counts.keys(), word_counts.values())

    wc = pd.DataFrame(temp, columns = ['word', 'count'])

    # ë‹¨ì–´ì˜ ìˆœìœ„
    # method='first': ê°™ì€ ê°’ì˜ ê²½ìš° ë¨¼ì €ë‚˜ì˜¨ ìš”ì†Œë¥¼ ìš°ì„ 
    wc['rank'] = wc['count'].rank(method='first', ascending=False)
    total = wc['count'].sum()

    # ì½”í¼ìŠ¤ ë‚´ ë‹¨ì–´ì˜ ë¹„ìœ¨
    wc['percent'] = wc['count'].apply(lambda x: x / total)

    wc = wc.sort_values(by='rank')

    # ëˆ„ì  ë¹„ìœ¨
    # cumsum() : cumulative sum
    wc['cul_percent'] = wc['percent'].cumsum()

    temp2 = zip(word_in_docs.keys(), word_in_docs.values())
    ac = pd.DataFrame(temp2, columns=['word', 'word_in_docs'])
    wc = ac.merge(wc, on='word')
    
    # ì „ì²´ ë¬¸ì„œ ì¤‘ ì¡´ì¬í•˜ëŠ” ë¹„ìœ¨
    wc['word_in_docs_percent'] = wc['word_in_docs'].apply(lambda x: x / total_docs)

    return wc.sort_values(by='rank')
```

### 4) SpaCyë¥¼ ì‚¬ìš©í•˜ì—¬ ë”ìš± ì‰½ê²Œ ì²˜ë¦¬í•˜ê¸°

`Spacy` ëŠ” ë¬¸ì„œ êµ¬ì„±ìš”ì†Œë¥¼ ë‹¤ì–‘í•œ êµ¬ì¡°ì— ë‚˜ëˆ„ì–´ ì €ì¥í•˜ëŠ” ëŒ€ì‹ , ìš”ì†Œë¥¼ ìƒ‰ì¸í™”í•˜ê³  ê²€ìƒ‰ì •ë³´ë¥¼ ê°„ë‹¨íˆ ì €ì¥í•˜ëŠ” ë¼ì´ë¸ŒëŸ¬ë¦¬ì…ë‹ˆë‹¤. <br/>
ê·¸ë ‡ê¸° ë•Œë¬¸ì— ì‹¤ì œ ë°°í¬ ë‹¨ê³„ì—ì„œ `NLTK` ë¼ì´ë¸ŒëŸ¬ë¦¬ë³´ë‹¤ `Spacy` ê°€ ë” ë¹ ë¦…ë‹ˆë‹¤.<br/>


```python
# spacy.io
import spacy
from spacy.tokenizer import Tokenizer

# Load general-purpose pretrained models to predict named entities, part-of-speech tags and syntactic dependencies
## https://spacy.io/models
## python -m spacy download en_core_web_sm
nlp = spacy.load("en_core_web_sm")

# Tokenizer ìƒì„±
tokenizer = Tokenizer(nlp.vocab)
```

`SpaCy` ë¥¼ ì‚¬ìš©í•˜ë©´ ë”ìš± ì‰½ê³  ë¹ ë¥´ê²Œ í† í°í™”ë¥¼ ìˆ˜í–‰í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```python
# Tokenizer Pipe

tokens = []

# í† í°í™” (in stream process)
# .pipe streams input, and produces streaming output
for doc in tokenizer.pipe(df['reviews.text']):
    doc_tokens = [token.text for token in doc]
    tokens.append(doc_tokens)

df['tokens'] = tokens
df['tokens'].head()
```

```python
wc = word_count(df['tokens'])
wc.head()
```

```python
wc_top20 = wc[wc['rank'] <= 20]

squarify.plot(sizes=wc_top20['percent'], label=wc_top20['word'], alpha=0.6 )
plt.axis('off')
plt.show()
```

### 5) ë¶ˆìš©ì–´(Stop words) ì²˜ë¦¬

```python
tokens = []
# í† í°ì—ì„œ ë¶ˆìš©ì–´ ì œê±°, ì†Œë¬¸ìí™” í•˜ì—¬ ì—…ë°ì´íŠ¸
for doc in tokenizer.pipe(df['reviews.text']):
    doc_tokens = []

    # A doc is a sequence of Token(<class 'spacy.tokens.doc.Doc'>)
    for token in doc:
        # í† í°ì´ ë¶ˆìš©ì–´ì™€ êµ¬ë‘ì ì´ ì•„ë‹ˆë©´ ì €ì¥
        if (token.is_stop == False) & (token.is_punct == False):
            doc_tokens.append(token.text.lower())

    tokens.append(doc_tokens)

df['tokens'] = tokens
df.tokens.head()
```

```python
wc = word_count(df['tokens'])

wc_top20 = wc[wc['rank'] <= 20]

squarify.plot(sizes=wc_top20['percent'], label=wc_top20['word'], alpha=0.6)
plt.axis('off')
plt.show()
```

```python
STOP_WORDS = nlp.Defaults.stop_words.union(['batteries','I', 'amazon', 'i', 'Amazon', 'it', "it's", 'it.', 'the', 'this'])
```

```python
tokens = []

for doc in tokenizer.pipe(df['reviews.text']):
    
    doc_tokens = []
    
    for token in doc: 
        if token.text.lower() not in STOP_WORDS:
            doc_tokens.append(token.text.lower())
   
    tokens.append(doc_tokens)
    
df['tokens'] = tokens
```

Stemming ì—ì„œëŠ” `wolf -> wolf`, `wolves -> wolv` ë¡œ ë³€í˜•ë˜ì—ˆì§€ë§Œ, <br/> Lemmatizationì—ì„œëŠ” `wolf -> wolf`, `wolves -> wolf` ë¡œ ë³€í˜•ë˜ëŠ” ê²ƒì„ ë³¼ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```python
# Lemmatization ê³¼ì •ì„ í•¨ìˆ˜ë¡œ ë§Œë“¤ì–´ ë´…ì‹œë‹¤
def get_lemmas(text):

    lemmas = []
    
    doc = nlp(text)

    for token in doc: 
        if ((token.is_stop == False) and (token.is_punct == False)) and (token.pos_ != 'PRON'):
            lemmas.append(token.lemma_)
    
    return lemmas
```

- **íŒŒë¼ë¯¸í„° íŠœë‹**

> ì´ë²ˆì—ëŠ” ì¡°ê¸ˆ ë” íŒŒë¼ë¯¸í„°ë¥¼ íŠœë‹í•˜ê³ , spacy tokenizer ë¥¼ ì‚¬ìš©í•´ì„œ ë²¡í„°í™”ë¥¼ ì§„í–‰í•´ ë³´ê² ìŠµë‹ˆë‹¤.

```python
# spacy tokenizer í•¨ìˆ˜
def tokenize(document):
    
    doc = nlp(document)
    # punctuations: !"#$%&'()*+,-./:;<=>?@[\]^_`{|}~
    return [token.lemma_.strip() for token in doc if (token.is_stop != True) and (token.is_punct != True) and (token.is_alpha == True)]
```

```python
# ngram_range = (min_n, max_n), min_n ê°œ~ max_n ê°œë¥¼ ê°–ëŠ” n-gram(nê°œì˜ ì—°ì†ì ì¸ í† í°)ì„ í† í°ìœ¼ë¡œ ì‚¬ìš©í•©ë‹ˆë‹¤.
# min_df = n, ìµœì†Œ nê°œì˜ ë¬¸ì„œì— ë‚˜íƒ€ë‚˜ëŠ” í† í°ë§Œ ì‚¬ìš©í•©ë‹ˆë‹¤
# max_df = .7, 70% ì´ìƒ ë¬¸ì„œì— ë‚˜íƒ€ë‚˜ëŠ” í† í°ì€ ì œê±°í•©ë‹ˆë‹¤
tfidf = TfidfVectorizer(stop_words='english'
                        ,tokenizer=tokenize
                        ,ngram_range=(1,2)
                        ,max_df=.7
                        ,min_df=3
#                         ,max_features = 4000
                       )

dtm = tfidf.fit_transform(data)
dtm = pd.DataFrame(dtm.todense(), columns=tfidf.get_feature_names())
dtm.head()
```
