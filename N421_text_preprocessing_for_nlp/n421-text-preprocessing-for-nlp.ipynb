ì‚¬ìš©í•  ì •ê·œì‹ í‘œí˜„ì€ `'[^a-zA-Z0-9 ]'` ì…ë‹ˆë‹¤. ì´ í‘œí˜„ì€ ì†Œë¬¸ì(a-z), ëŒ€ë¬¸ì(A-Z), ìˆ«ì(0-9), ê³µë°±ë¬¸ì(space)ë¥¼ ì œì™¸í•œ ëª¨ë“  ë¬¸ìë¥¼ ì œê±°í•©ë‹ˆë‹¤.

- [ì°¸ê³ : Python RegEx](https://www.w3schools.com/python/python_regex.asp#sub)
- [ì •ê·œ í‘œí˜„ì‹ ì‹œì‘í•˜ê¸°](https://wikidocs.net/4308#_2)

````python
# ì •ê·œì‹ ë¼ì´ë¸ŒëŸ¬ë¦¬
import re

# ì •ê·œì‹
# []: [] ì‚¬ì´ ë¬¸ìë¥¼ ë§¤ì¹˜, ^: not
regex = r"[^a-zA-Z0-9 ]"

# ì •ê·œì‹ì„ ì ìš©í•  ìŠ¤íŠ¸ë§
test_str = ("(Natural Language Processing) is easy!, DS!\n")

# ì¹˜í™˜í•  ë¬¸ì
subst = ""

result = re.sub(regex, subst, test_str)
````

#### ì •ê·œì‹ ì •ì˜ì—ì„œ ì¹˜í™˜ê¹Œì§€ ìˆ˜í–‰í•˜ëŠ” tokenize í•¨ìˆ˜ë¥¼ ë§Œë“¤ì–´ ë³´ì„¸ìš”.

````python
def tokenize(text):
    """text ë¬¸ìì—´ì„ ì˜ë¯¸ìˆëŠ” ë‹¨ì–´ ë‹¨ìœ„ë¡œ listì— ì €ì¥í•©ë‹ˆë‹¤.
    Args:
        text (str): í† í°í™” í•  ë¬¸ìì—´
    Returns:
        list: í† í°ì´ ì €ì¥ëœ ë¦¬ìŠ¤íŠ¸
    """
    # ì •ê·œì‹ ì ìš©
    tokens = re.sub(regex, subst, text)

    # ì†Œë¬¸ìë¡œ ì¹˜í™˜
    tokens = tokens.lower().split()
    
    return tokens
````

the, and, a, is... ê°™ì€ ì–´ë–¤ ë¬¸ì„œì—ë“ ì§€ ë§ì´ ë‚˜íƒ€ë‚˜ëŠ” ë‹¨ì–´ë“¤ì€ ì‚¬ì‹¤ í° ì˜ë¯¸ë¥¼ ì°¾ê¸° ì–´ë ¤ìš´ ë‹¨ì–´ë“¤ ì…ë‹ˆë‹¤.

ì´ëŸ° ë‹¨ì–´ë“¤ì„ **stopword(ë¶ˆìš©ì–´)** ë¼ê³  í•©ë‹ˆë‹¤.

ì´ì™€ê°™ì€ í”„ë¡œì„¸ìŠ¤ë¥¼ ì‚¬ìš©í•´ì„œ ì½”í¼ìŠ¤ì˜ ì „ì²´ ì›Œë“œ ì¹´ìš´íŠ¸, ë­í¬ ë“± ì •ë³´ê°€ ë‹´ê¸´ ë°ì´í„°í”„ë ˆì„ì„ ë¦¬í„´í•˜ëŠ” í•¨ìˆ˜ë¥¼ ë§Œë“¤ì–´ ë´…ì‹œë‹¤. ì´ í•¨ìˆ˜ëŠ” í† í°í™”ëœ ë¬¸ì„œë“¤ì´ ë“¤ì–´ìˆëŠ” ì½”í¼ìŠ¤ë¥¼ ì…ë ¥ìœ¼ë¡œ ë°›ìŠµë‹ˆë‹¤.

````python

def word_count(docs):
    """ í† í°í™”ëœ ë¬¸ì„œë“¤ì„ ì…ë ¥ë°›ì•„ í† í°ì„ ì¹´ìš´íŠ¸ í•˜ê³  ê´€ë ¨ëœ ì†ì„±ì„ ê°€ì§„ ë°ì´í„°í”„ë ˆì„ì„ ë¦¬í„´í•©ë‹ˆë‹¤.
    Args:
        docs (series or list): í† í°í™”ëœ ë¬¸ì„œê°€ ë“¤ì–´ìˆëŠ” list
    Returns:
        list: Dataframe
    """
    # ì „ì²´ ì½”í¼ìŠ¤ì—ì„œ ë‹¨ì–´ ë¹ˆë„ ì¹´ìš´íŠ¸
    word_counts = Counter()

    # ë‹¨ì–´ê°€ ì¡´ì¬í•˜ëŠ” ë¬¸ì„œì˜ ë¹ˆë„ ì¹´ìš´íŠ¸, ë‹¨ì–´ê°€ í•œ ë²ˆ ì´ìƒ ì¡´ì¬í•˜ë©´ +1
    word_in_docs = Counter()

    # ì „ì²´ ë¬¸ì„œì˜ ê°¯ìˆ˜
    total_docs = len(docs)

    for doc in docs:
        word_counts.update(doc)
        word_in_docs.update(set(doc))

    temp = zip(word_counts.keys(), word_counts.values())

    wc = pd.DataFrame(temp, columns = ['word', 'count'])

    # ë‹¨ì–´ì˜ ìˆœìœ„
    # method='first': ê°™ì€ ê°’ì˜ ê²½ìš° ë¨¼ì €ë‚˜ì˜¨ ìš”ì†Œë¥¼ ìš°ì„ 
    wc['rank'] = wc['count'].rank(method='first', ascending=False)
    total = wc['count'].sum()

    # ì½”í¼ìŠ¤ ë‚´ ë‹¨ì–´ì˜ ë¹„ìœ¨
    wc['percent'] = wc['count'].apply(lambda x: x / total)

    wc = wc.sort_values(by='rank')

    # ëˆ„ì  ë¹„ìœ¨
    # cumsum() : cumulative sum
    wc['cul_percent'] = wc['percent'].cumsum()

    temp2 = zip(word_in_docs.keys(), word_in_docs.values())
    ac = pd.DataFrame(temp2, columns=['word', 'word_in_docs'])
    wc = ac.merge(wc, on='word')
    
    # ì „ì²´ ë¬¸ì„œ ì¤‘ ì¡´ì¬í•˜ëŠ” ë¹„ìœ¨
    wc['word_in_docs_percent'] = wc['word_in_docs'].apply(lambda x: x / total_docs)

    return wc.sort_values(by='rank')
````

````python
lem = "The social wolf. Wolves are complex."

nlp = spacy.load("en_core_web_sm")

doc = nlp(lem)

# wolf, wolveê°€ ì–´ë–¤ Lemmaë¡œ ì¶”ì¶œë˜ëŠ”ì§€ í™•ì¸í•´ ë³´ì„¸ìš”
for token in doc:
    print(token.text, "  ", token.lemma_)
````

````python
# Lemmatization ê³¼ì •ì„ í•¨ìˆ˜ë¡œ ë§Œë“¤ì–´ ë´…ì‹œë‹¤
def get_lemmas(text):

    lemmas = []
    
    doc = nlp(text)

    for token in doc: 
        if ((token.is_stop == False) and (token.is_punct == False)) and (token.pos_ != 'PRON'):
            lemmas.append(token.lemma_)
    
    return lemmas
````

````python
df['lemmas'] = df['reviews.text'].apply(get_lemmas)
````
