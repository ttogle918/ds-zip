```python
# ì´ ê³³ì— ë‹µì•ˆì„ ì‘ì„±í•˜ì‹œê¸° ë°”ëë‹ˆë‹¤
import pandas as pd
trees = pd.read_csv('https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/seoul_tree/seoul_tree.txt', sep = '\t', skiprows = 1)

trees = trees.replace({'-':0})

trees.head()
```

```python
 trees = trees[~trees['ìì¹˜êµ¬'].isin(['ì„œìš¸ì‹œ', 'ê³µì›ë…¹ì§€ì‚¬ì—…ì†Œ','ì‹œì„¤ê´€ë¦¬ê³µë‹¨'])]
# trees =trees[(trees['ìì¹˜êµ¬'] != 'ì„œìš¸ì‹œ') & (trees['ìì¹˜êµ¬'] != 'ê³µì›ë…¹ì§€ì‚¬ì—…ì†Œ') & (trees['ìì¹˜êµ¬'] != 'ì‹œì„¤ê´€ë¦¬ê³µë‹¨')]
trees.head()
```

## 2. Two-Sample T-test ì—°ìŠµ

### "ì„œìš¸ì‹œì˜ êµ¬ë³„ í‰ê·  ëŠí‹°ë‚˜ë¬´ìˆ˜ì™€ ì™•ë²šë‚˜ë¬´ì˜ ìˆ˜ëŠ” ì°¨ì´ê°€ ì—†ë‹¤"ë¼ëŠ” ê°€ì„¤ì— ëŒ€í•´ ê°€ì„¤ê²€ì •ì„ ì‹œí–‰í•˜ì„¸ìš”.

- ë‹¨, `ì„œìš¸ì‹œ`, `ê³µì›ë…¹ì§€ì‚¬ì—…ì†Œ`, `ì‹œì„¤ê´€ë¦¬ê³µë‹¨` ë¶€ë¶„ì˜ ë°ì´í„°ëŠ” ì œì™¸í•©ë‹ˆë‹¤.
- ê°€ì„¤ê²€ì • ê²°ê³¼ì˜ pvalueë¥¼ `pv3`ì— ì €ì¥í•˜ì„¸ìš”

```python
# ê°€ì„¤ ê²€ì •
# H0(ê·€ë¬´ê°€ì„¤) : ëŠí‹°ë‚˜ë¬´, ì™•ë²šë‚˜ë¬´ ìˆ˜ì˜ ê°ê°ì˜ í‰ê· ì€ ì°¨ì´ê°€ ì—†ë‹¤.
# H1(ëŒ€ë¦½ê°€ì„¤) : ëŠí‹°ë‚˜ë¬´, ì™•ë²šë‚˜ë¬´ ìˆ˜ì˜ ê°ê°ì˜ í‰ê· ì€ ì°¨ì´ê°€ ìˆë‹¤.

tree2 = pd.to_numeric(trees['ì™•ë²šë‚˜ë¬´'].str.replace(',',''))
tree3 = pd.to_numeric(trees['ëŠí‹°ë‚˜ë¬´'].str.replace(',',''))

print("ì™•ë²šë‚˜ë¬´ mean:", np.mean(tree2))
print("ëŠí‹°ë‚˜ë¬´ mean:", np.mean(tree3))
```

```python
# p-valueê°’ì´ ìœ ì˜ìˆ˜ì¤€(0.05)ë³´ë‹¤ ë†’ìŒ -> ê·€ë¬´ê°€ì„¤ ê¸°ê° ë¶ˆê°€

sta, pv3 = stats.ttest_ind(tree2, tree3)

print(sta, pv3)
```
