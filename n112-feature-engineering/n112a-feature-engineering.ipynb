Relative Perfomance ë¼ëŠ” ìƒˆë¡œìš´ featureë¥¼ ê³„ì‚°í•˜ì„¸ìš”.

```
mean = df.iloc[1:, 1].sum() / 4
# meanì„ êµ¬í•˜ê³  meanë³€ìˆ˜ì— ì €ì¥

grades = []
for row in df['ë§¤ì¶œì•¡']:
    if row >= (mean*1.1):
            grades.append('S')
    elif row >= (mean*1.05):
            grades.append('A')
    elif row >= (mean*0.95):
            grades.append('B')
    elif row >= (mean*0.90):
            grades.append('C')
    else:
            grades.append('D');
df['ë“±ê¸‰'] = grades

# S: 13798.4 ì´ìƒ
# A: 13171.2
# B: 11916.8
# C: 11289.6
# D: 11289.6 ì´í•˜

df
```
