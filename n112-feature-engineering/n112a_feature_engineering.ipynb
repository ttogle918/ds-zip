```python
import pandas as pd
from pandas import Series, DataFrame

url = 'https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/kt%26g/kt%26g_0.csv'
headers = ['분기', '매출액', '영업이익', '영업이익(발표기준)', '세전계속사업이익',
           '당기순이익', '당기순이익(지배)', '당기순이익(비지배)', '자산총계', '부채총계',
           '자본총계', '자본총계(지배)', '자본총계(비지배)', '자본금', '영업활동현금흐름',
           '투자활동현금흐름', '재무활동현금흐름', '영업이익률', '순이익률', 'ROE(%)',
           'ROA(%)', '부채비율', '자본유보율', 'EPS(원)', 'PER(배)']
def toInt(str1):
  if type(str1) == int or type(str1) == float:
    return str1
  return int(str1.replace(',', ''))

def toFloat(str1):
  if type(str1) == float :
    return str1
  return float(str1.replace(',', ''))
df = pd.read_csv(url, names=headers)

# 영업이익률 = 영업이익/매출액 * 100
df['영업이익률2'] = df['영업이익'].apply(toInt) / df['매출액'].apply(toInt) * 100

# TAM : 전체 시장 규모
# PDR = 기업가치(시가총액) / (TAM * 시장점유율)
# 기업가치(EV) = 시가총액(MC) + 총부채(Total Debt) - 현금 및 현금성 자산(C)
# 시가총액 = 자본총계
df2 = DataFrame({'구분': ['18년', '19년'], '종가': [101500, 93800], '발행주식수': [137292497, 137292497], 
                 '시가총액': ['13935188445500', '12878036218600'], '시장점유율':[62.0, 63.5], '시장의규모' :[15000000000000, 15000000000000]})


df2['PDR'] = df2['시가총액'].apply(toInt) / ( df2['시장의규모'] + df2['시장점유율']) 
# data를 정수(숫자)로 입력했는데, Int, float로.. type캐스팅 할려고 하니까 계속 에러가 났음.
df
```
