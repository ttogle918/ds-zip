```python
merged = (products[['product_id', 'product_name']]
          .merge(order_products[['order_id', 'product_id']])
          .merge(orders[['order_id', 'order_hour_of_day']]))

merged.head()
```

ì‚¬ì‹¤ ìœ„ì—ì„œ í•œ ì‘ì—…ì€ ëª¨ë“  ì œí’ˆì— ëŒ€í•´ì„œ ì´ë¤„ì§„ ê²ƒìœ¼ë¡œ ìš°ë¦¬ê°€ í•„ìš”ë¡œ í•˜ëŠ” 2ê°œì˜ ì œí’ˆì— ëŒ€í•´ì„œëŠ” ì„ íƒí›„, ë³‘í•©ì´ë¼ëŠ” ë‹¨ê³„ë¥¼ ê±°ì³ì•¼í•©ë‹ˆë‹¤. 
- ë³‘í•© í›„ ì„ íƒ 
- ì„ íƒ í›„ ë³‘í•© 
ë‘ê°€ì§€ ë°©ë²• ëª¨ë‘ ê²°ê³¼ë¬¼ì€ ê°™ì§€ë§Œ ì°¨ì´ëŠ” ë¶„ëª…íˆ ì¡´ì¬í•©ë‹ˆë‹¤. ë‹¤ë§Œ ì§€ê¸ˆì€ ê·¸ ì°¨ì´ì— ì§‘ì¤‘í•˜ëŠ” ê²ƒë³´ë‹¨, ê²°ê³¼ë¬¼ì„ í™•ì¸í•˜ëŠ”ê²ƒì„ ìš°ì„ ìœ¼ë¡œ í•˜ë„ë¡ í•˜ê² ìŠµë‹ˆë‹¤.

```python
condition = ((merged['product_name']=='Half Baked Frozen Yogurt') | 
             (merged['product_name']=='Half And Half Ultra Pasteurized'))

merged = merged[condition]

print(merged.shape)
merged.head()
```

ì•ì—ì„œì²˜ëŸ¼, ì»¨ë””ì…˜ì„ ì„¤ì •í•˜ëŠ”ë°ì— ìˆì–´ì„œëŠ” ì—¬ëŸ¬ê°€ì§€ ë°©ë²•ì´ ìˆìŠµë‹ˆë‹¤. 

```python
product_names = ['Half Baked Frozen Yogurt', 'Half And Half Ultra Pasteurized']

condition = merged['product_name'].isin(product_names)

subset = merged[condition]

print(subset.shape)
subset.head()
```

```python
froyo = subset[subset['product_name']=='Half Baked Frozen Yogurt']
cream = subset[subset['product_name']=='Half And Half Ultra Pasteurized']
```

4) Melt 

ìš°ì„  ë°ì´í„°ë¥¼ wide í•œ í˜•íƒœë¡œ ë°”ê¾¸ì–´ì•¼ í•©ë‹ˆë‹¤. crosstab ë˜í•œ pivot_tableì˜ ì¼ì¢…ì´ë‹ˆ ì´ë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤.

```python
table = pd.crosstab(subset['order_hour_of_day'], 
                    subset['product_name'], 
                    normalize=True)

table
```

```python
melted = table.reset_index().melt(id_vars='order_hour_of_day').rename(columns={
     'order_hour_of_day': 'Hour of Day Ordered', 
     'product_name': 'Product', 
     'value': 'Percent of Orders by Product'
 })

melted
```

```python
sns.relplot(x='Hour of Day Ordered', 
            y='Percent of Orders by Product', 
            hue='Product', 
            data=melted, 
            kind='line');
```
