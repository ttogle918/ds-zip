```python
import seaborn as sns
import pandas as pd
import numpy as np

tips = sns.load_dataset('tips')
tips.head()
```

## Violin plot 

boxplot ÎåÄÏã†Ïóê density Î•º Í∑∏Î¶∞ Ï∞®Ìä∏.

<https://seaborn.pydata.org/generated/seaborn.violinplot.html#seaborn.violinplot>

```python
# Time / TipÏùò Í¥ÄÍ≥Ñ

sns.violinplot(
    x = 'time', 
    y = 'tip',
    data = tips
);
```

```python
sns.violinplot(
    x = 'smoker',
    y = 'tip',
    data = tips
);
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips);
```

## Barplot

```python
sns.barplot(x = 'day', y = 'tip', data = tips);
```

```python
sns.barplot(y = 'day', x = 'tip', data = tips);
```

## ÏÉâÏÉÅ Palette 

```python
sns.barplot(y = 'day', x = 'tip', data = tips, palette = sns.cubehelix_palette(len(tips['day'].unique())));
```

## Label and Annotation

```python
import matplotlib.pyplot as plt

sns.set_style('whitegrid') #set theme
fig,ax = plt.subplots()

chart = sns.barplot(y = 'day', x = 'tip', data = tips, palette = sns.cubehelix_palette(len(tips['day'].unique())));
chart.set_xlabel('Amount of Tip', weight = 'bold', fontsize = 13)
chart.set_ylabel('Days', weight = 'bold', fontsize = 13)
chart.set_title('Tip for each Days', weight = 'bold', fontsize = 16)

ax.text(0, 0, 'TEXT 0, 0');
ax.text(0, 1, 'TEXT 0, 1');
ax.text(1, 0, 'TEXT 1, 0');
ax.text(1, 1, 'TEXT 1, 1');
```

## CountPlot

```python
sns.countplot(data = tips, x = 'day');
```

# Grouping (facet)

```python
g = sns.FacetGrid(data = tips, col = 'time')

g.map_dataframe(
    sns.scatterplot,
    x = 'total_bill',
    y = 'tip',
    hue = 'day',
    alpha = 0.7, # tranparency
    data = tips
)
g.add_legend();
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', hue = 'sex', data = tips);
```

```python
g = sns.FacetGrid(data = tips, col = 'smoker') # colÏù¥ ÏïÑÎãàÎùº row Î°úÎèÑ Ïã§ÌñâÌï¥Î≥¥ÏÑ∏Ïöî.

g.map_dataframe(
    sns.scatterplot,
    x = 'total_bill',
    y = 'tip',
    hue = 'sex',
    alpha = 0.5, # tranparency
    data = tips
)
g.add_legend();
```

# 2d facet

```python
g = sns.FacetGrid(data = tips, row = 'smoker', col = 'day')

g.map_dataframe(
    sns.scatterplot,
    x = 'total_bill',
    y = 'tip',
    hue = 'sex', # color
    alpha = 0.5, # tranparency
    data = tips
)
g.add_legend();
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips);
```

### ÏÉâÏÉÅ palette ÏßÄÏ†ï ( dictionary ÏÇ¨Ïö© )

```python
sns.boxplot(x = 'time', y = 'tip', data = tips, hue = 'time', palette = {"Dinner": '#ff9ff3', "Lunch": '#1dd1a1'});
```

```python
g = sns.FacetGrid(data = tips, col = 'sex')

g.map_dataframe(
    sns.violinplot,
    x = 'day',
    y = 'tip',
    hue = 'sex',
    data = tips, 
    palette = {'Male':'#f1c40f', 'Female':'#d35400'}
)
g.add_legend();
```
