```python
from scipy import stats

# ttest_1samp í•¨ìˆ˜ì˜ íŒŒë¼ë¯¸í„° 1) Sample ë°ì´í„°, 2) ë¹„êµí•˜ë ¤ëŠ” ê°’

stats.ttest_1samp(coinflips, .5)
```

### One-side test vs Two-side test

Two side (tail / direction) test : ìƒ˜í”Œ ë°ì´í„°ì˜ í‰ê· ì´ "X"ì™€ ê°™ë‹¤ / ê°™ì§€ ì•Šë‹¤. ë¥¼ ê²€ì •í•˜ëŠ” ë‚´ìš©

One side test : ìƒ˜í”Œ ë°ì´í„°ì˜ í‰ê· ì´ "X"ë³´ë‹¤ í¬ë‹¤ í˜¹ì€ ì‘ë‹¤ / í¬ì§€ ì•Šë‹¤ ì‘ì§€ ì•Šë‹¤. ë¥¼ ê²€ì •í•˜ëŠ” ë‚´ìš©

### Two Sample T-test

2ê°œì˜ sample ê°’ë“¤ì˜ **í‰ê· **ì´ **ì„œë¡œ** ë™ì¼í•œì§€ ë¹„êµ.

2ê°œì˜ ë™ì „ (500ì›ì§œë¦¬ vs 100ì›ì§œë¦¬)ì„ ì—¬ëŸ¬ë²ˆ ë˜ì ¸ì„œ p(H)ì˜ í‰ê· ì´ ìœ ì‚¬í•œ ì§€ 

1) ê·€ë¬´ê°€ì„¤ : ë‘ í™•ë¥ ì€ ê°™ë‹¤ (ì°¨ì´ê°€ ì—†ë‹¤).

$H_0: \bar{x}_1 = \bar{x}_2$

2) ëŒ€ì•ˆê°€ì„¤ : ê°™ì§€ ì•Šë‹¤

$H_1: \bar{x}_1 \neq \bar{x}_2$

3) ì‹ ë¢°ë„ : 95%



```python
np.random.seed(111)

coin1 = np.random.binomial(n = 1, p = 0.5, size = 500)
coin2 = np.random.binomial(n = 1, p = 0.5, size = 200)

print(np.mean(coin1))
print(np.mean(coin2))

stats.ttest_ind(coin1, coin2) 
```

```python
np.random.seed(111)

coin1 = np.random.binomial(n = 1, p = 0.6, size = 500)
coin2 = np.random.binomial(n = 1, p = 0.5, size = 200)

print(np.mean(coin1))
print(np.mean(coin2))

stats.ttest_ind(coin1, coin2) 
```
