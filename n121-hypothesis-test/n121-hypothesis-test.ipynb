## The T-test Process:

1) ê·€ë¬´ ê°€ì„¤ (Null Hypothesis) ë¥¼ ì„¤ì • (ë™ì „ì€ ê³µì •í•˜ë‹¤)

$H_0: \mu = \bar{x}$

$\mu =$ ëª¨ì§‘ë‹¨ì˜ í‰ê· 

$\bar{x} =$ í‘œë³¸ì˜ í‰ê· 

2) ëŒ€ì•ˆ ê°€ì„¤ (Alternative Hypothesis) ë¥¼ ì„¤ì • (ë™ì „ì€ ê³µì •í•˜ì§€ ì•Šë‹¤)

$H_1: \mu \neq \bar{x}$

3) ì‹ ë¢°ë„ë¥¼ ì„¤ì • (Confidence Level) : ëª¨ìˆ˜ê°€ ì‹ ë¢° êµ¬ê°„ ì•ˆì— í¬í•¨ë  í™•ë¥  (ë³´í†µ 95, 99% ë“±ì„ ì‚¬ìš©)

ì‹ ë¢°ë„ 95%ì˜ ì˜ë¯¸ 
= ëª¨ìˆ˜ê°€ ì‹ ë¢° êµ¬ê°„ ì•ˆì— í¬í•¨ë  í™•ë¥ ì´ 95%
= ê·€ë¬´ê°€ì„¤ì´ í‹€ë ¸ì§€ë§Œ ìš°ì—°íˆ ì„±ë¦½í•  í™•ë¥ ì´ 5%

4) P-valueë¥¼ í™•ì¸

```python
from scipy import stats
# íŒŒë¼ë¯¸í„° 1) Sample ë°ì´í„°, 2) ë¹„êµí•˜ë ¤ëŠ” ê°’
coinflips = np.random.binomial(n = 1, p = 0.5, size = 1000)
print(stats.ttest_1samp(coinflips, .5))

coinflips = np.random.binomial(n = 1, p = 0.8, size = 1000)
print(np.mean(coinflips))
print(stats.ttest_1samp(coinflips, .5))
print(stats.ttest_1samp(coinflips, .8))
```

ë°ì´í„°ëŠ” ê°ê° 5 (í¸ì°¨ 3.2), 7(2.1) kgë¼ê³  í•  ë•Œ ë‘ ìˆ«ìëŠ” ë‹¤ë¥´ì§€ë§Œ, ê³ ì–‘ì´ì™€ ê°•ì•„ì§€ì˜ ëª¸ë¬´ê²Œê°€ í†µê³„ì ìœ¼ë¡œ ìœ ì˜í•œ ì°¨ì´ê°€ ìˆëŠ”ì§€ ì—†ëŠ”ì§€ì— ëŒ€í•˜ì—¬ ê²°ë¡ ì„ ë‚´ë ¤ë©´ ì–´ë–»ê²Œ í•´ì•¼ í• ê¹Œìš”?

$H_0: \bar{x}_1 = \bar{x}_2$

2) ëŒ€ì•ˆê°€ì„¤ : ê°™ì§€ ì•Šë‹¤

```python
xbar1 = 5
s1 = 3.2

xbar2 = 7
s2 = 2.1

sample1 = np.random.normal(xbar1, s1, 1000) # mean, sd, size
sample2 = np.random.normal(xbar2, s2, 800) # mean, sd, size

stats.ttest_ind(sample1, sample2)
```

```python
import pandas as pd

trees = pd.read_csv('https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/seoul_tree/seoul_tree.txt', sep = '\t', skiprows = 1)

trees = trees.replace({'-':0})

print(trees.head())
```

```python
tree1 = pd.to_numeric(trees['ì€í–‰ë‚˜ë¬´'].str.replace(',',''))
tree2 = pd.to_numeric(trees['ëŠí‹°ë‚˜ë¬´'].str.replace(',',''))

print(tree1.head())
print(tree2.head())
```

## 1-sample t-test: ì€í–‰ë‚˜ë¬´ì˜ í‰ê·

```python
print(np.mean(tree1))

print(stats.ttest_1samp(tree1, 20000))
print(stats.ttest_1samp(tree1, 8000))
print(stats.ttest_1samp(tree1, 7800))
print(stats.ttest_1samp(tree1, 5000))
print(stats.ttest_1samp(tree1, 4000))
```
