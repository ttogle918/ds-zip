## Run a $\chi^{2}$ Test (Numpyë¡œ ì§ì ‘ ê³„ì‚°)

ê°€ëŠ¥í•œ 3ê°€ì§€ì˜ ì¡°í•© ì¤‘ 2ê°€ì§€ì˜ ê²½ìš°ì— ëŒ€í•´ì„œë§Œ í•´ë³´ê² ìŠµë‹ˆë‹¤.

1) `income` & `pay_meal_out`

2) `pay_meal_out` & `persian_food`


```python
obs = pd.crosstab(small_food['income'], small_food['pay_meal_out'])
obs
```

- Numpyë¡œ ê³„ì‚°í•˜ê¸°ì—ëŠ” categoryê°€ ë„ˆë¬´ ë§ìœ¼ë‹ˆ 3, 4 ë°ì´í„°ë§Œ ì‚¬ìš©í•˜ë„ë¡ í•˜ê² ìŠµë‹ˆë‹¤.

| |pay = 3|pay = 4|
|:-:|:-:|:-:|
|income = 3|9|2|
|income = 4|13|3|


### ì˜ˆì¸¡ê°’ ê³„ì‚°
\begin{align}
expected_{i,j} =\frac{(row_{i} \text{total})(column_{j} \text{total}) }{(\text{total observations})}  
\end{align}

í•´ë‹¹ ë‚´ìš©ì— ëŒ€í•œ ì¶”ê°€ ì„¤ëª…ê³¼ ì˜ˆì‹œëŠ” urclassë¥¼ ( **2 Sample chi-square Expected Value** ) ì°¸ì¡°í•˜ì„¸ìš”.

```python
obs = np.array([[9, 2], [13, 3]])
obs
```

```python
total_obs = np.sum(obs)
total_obs
```

```python
exp = np.array([[(9+2)*(9+13), (9+2)*(2+3)], [(13+3)*(9+13), (13+3)*(2+3)]])

exp = exp / total_obs
exp
```

### $\chi^{2}$ í†µê³„ì¹˜ ê³„ì‚°

\begin{align}
\chi^2 = \sum \frac{(observed_i-expected_i)^2}{(expected_i)}
\end{align}

```python
squared = np.power(obs - exp, 2)
squared
```

```python
x2 = np.sum(squared / exp)
x2
```

2-sampe $\chi^2$ testì˜ ììœ ë„ (Degree of Freedom) ëŠ” 1-sampleê³¼ëŠ” ì¡°ê¸ˆ ë‹¤ë¦…ë‹ˆë‹¤.

1-sample (ì í•©ë„ ê²€ì •), DF = # categories-1

2-sample (ë…ë¦½ì„± ê²€ì •), DF = (#í–‰ - 1)*(#ì—´ - 1)

DF: (2-1)*(2-1) = 1

## Scipyë¥¼ ì´ìš©í•œ $\chi^{2}$ í…ŒìŠ¤íŠ¸

1) ê·€ë¬´ê°€ì„¤ : ë‘ ë³€ìˆ˜ê°„ ì—°ê´€ì´ ì—†ë‹¤.

2) ëŒ€ë¦½ê°€ì„¤ : ë‘ ë³€ìˆ˜ê°„ ì—°ê´€ì´ ìˆë‹¤.

3) ì‹ ë¢°êµ¬ê°„ : 95% 

```python
## 2 sample chi square test

# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.chi2_contingency.html

print(exp)

from scipy.stats import chi2_contingency

print('---')
print(chi2_contingency(obs)) # correction íŒŒë¼ë¯¸í„°ê°€ Trueë¡œ ì„¤ì • :  Yatesâ€™ correction ì‹œí–‰í•¨ : pvalueëŠ” ì•½ê°„ ë‹¤ë¦„

print('---')
print(chi2_contingency(obs, correction = False)) # ìœ„ì—ì„œ ê³„ì‚°í•œ ê²ƒê³¼ ë™ì¼
```
