## Run a $\chi^{2}$ Test "by hand" (Using Numpy)

ê°€ëŠ¥í•œ 3ê°€ì§€ì˜ ì¡°í•© ì¤‘ 2ê°€ì§€ì˜ ê²½ìš°ì— ëŒ€í•´ì„œë§Œ í•´ë³´ê² ìŠµë‹ˆë‹¤.

1) income - pay_meal_out

2) pay_meal_out - persian_food


```python
obs = pd.crosstab(small_food['income'], small_food['pay_meal_out'])
print(obs)

print(obs.sum()) # sum by column
```

## ì²˜ìŒì— ê³„ì‚°í•˜ê¸°ì— categoryê°€ ë„ˆë¬´ ë§ìœ¼ë‹ˆ 3,4 ë°ì´í„°ë§Œ ì‚¬ìš©í•˜ë„ë¡ í•˜ê² ìŠµë‹ˆë‹¤.

||pay = 3|pay = 4|
|---|---|---|
|income = 3|9|2 |
|income = 4| 13|3 |


## ì˜ˆì¸¡ê°’ ê³„ì‚°
\begin{align}
expected_{i,j} =\frac{(row_{i} \text{total})(column_{j} \text{total}) }{(\text{total observations})}  
\end{align}

```python
obs = np.array([9,2,13,3])
```

```python
total_obs = np.sum(obs)
print(total_obs)
```

```python
exp = np.array([(9+13)*(9+2), (9+13)*(13+3), (2+3)*(9+2), (2+3)*(13+3)])
```

### $\chi^{2}$ í†µê³„ì¹˜ ê³„ì‚°

\begin{align}
\chi^2 = \sum \frac{(observed_i-expected_i)^2}{(expected_i)}
\end{align}

```python
squared = np.power(obs - exp, 2)
print(squared)
```

```python
x2 = np.sum(squared / exp)
print(x2)
```

2-sampe $\chi^2$ testì˜ ììœ ë„ (Degree of Freedom) ëŠ” 1-sampleê³¼ëŠ” ì¡°ê¸ˆ ë‹¤ë¦…ë‹ˆë‹¤.

1-sample (ì í•©ë„ ê²€ì •), DF = # categories-1

2-sample (ë…ë¦½ì„± ê²€ì •), DF = (#í–‰ - 1)*(#ì—´ - 1)

DF: (2-1)*(2-1) = 1

P-value ê³„ì‚°ì„ ìœ„í•œ í†µê³„ì¹˜ ì‚¬ìš© : 

stats.chi2.pdf(x2, df)

[U Iowa chi^2 applet](https://homepage.divms.uiowa.edu/~mbognar/applets/chisq.html)

```python
from scipy import stats
print( 1 - stats.chi2.cdf(x2, df = 1)) # pvalue : 0.0, ì—°ê´€ì´ ìˆë‹¤.
```

## Scipyë¥¼ ì´ìš©í•œ $\chi^{2}$ í…ŒìŠ¤íŠ¸

3) ì‹ ë¢°êµ¬ê°„ : 95% 

```python
# ?stats.chisquare

chisquare(obs, exp, ddof = 1, axis = None)
```
