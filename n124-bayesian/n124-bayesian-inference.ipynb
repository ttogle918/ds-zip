## ë² ì´ì§€ì•ˆ ì˜ˆì œ 

ë‹¤ì‹œ í•œë²ˆ ì •ë¦¬í•˜ìë©´, ë² ì´ì§€ì•ˆì—ì„œì˜ $P(A|B)$ëŠ” $B$ë¼ëŠ” ì¦ê±°ë¥¼ ë°”íƒ•ìœ¼ë¡œí•œ $A$ì˜ í™•ë¥ ì…ë‹ˆë‹¤.

- TPR : True Positive Rate (=ë¯¼ê°ë„, true accept rate)
1ì¸ ì¼€ì´ìŠ¤ì— ëŒ€í•´ 1ë¡œ ì˜ ì˜ˆì¸¡í•œ ë¹„ìœ¨.(ì•”í™˜ìë¥¼ ì§„ì°°í•´ì„œ ì•”ì´ë¼ê³  ì§„ë‹¨ í•¨)

- FPR :  False Positive Rate (=1-íŠ¹ì´ë„, false accept rate)
0ì¸ ì¼€ì´ìŠ¤ì— ëŒ€í•´ 1ë¡œ ì˜ëª» ì˜ˆì¸¡í•œ ë¹„ìœ¨.(ì•”í™˜ìê°€ ì•„ë‹Œë° ì•”ì´ë¼ê³  ì§„ë‹¨ í•¨)

```python
# ë² ì´ì§€ì•ˆ ê³„ì‚°ì„ ìœ„í•´ì„œëŠ” 4ê°œì˜ ë³€ìˆ˜ê°€ í•„ìš”í•©ë‹ˆë‹¤.

p_pos_used = 0.99 # True positive rate (TPR, Sensitivity)
p_used = 0.005 # prior probability
p_pos_not_used = 0.01 # False positive rate (FPR)
p_not_used = 1 - p_used # 1 - p_used  

numerator = p_pos_used * p_used 

denominator = (p_pos_used * p_used) + (p_pos_not_used * p_not_used)

posterior_probability = numerator / denominator

posterior_probability
```

```python
import numpy as np

from scipy import stats

coinflips = np.random.binomial(1, 0.5, 20)
coinflips
```

```python
# Frequency ì‹ ë¢° êµ¬ê°„ ( module 3ì˜ ì½”ë“œ )

def confidence_interval(data, confidence=0.95):
  
    """
      ì£¼ì–´ì§„ ë°ì´í„°ì˜ í‘œë³¸ í‰ê· ì— ëŒ€í•œ ì‹ ë¢° êµ¬ê°„ì„ ê³„ì‚°í•©ë‹ˆë‹¤.
      ê¸°ë³¸ ê°’ìœ¼ë¡œ t-ë¶„í¬ì™€ ì–‘ë°©í–¥ (two-tailed), 95%ì˜ ì‹ ë¢°ë„ë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤. 
  
      ì…ë ¥ ê°’ : 
        data - ì—¬ëŸ¬ ê°œë¡œ ì´ë£¨ì–´ì§„ (list í˜¹ì€ numpy ë°°ì—´) í‘œë³¸ ê´€ì¸¡ì¹˜
        confidence - ì‹ ë¢°êµ¬ê°„ì„ ìœ„í•œ ì‹ ë¢°ë„ 
  x`x`
      ë°˜í™˜ ë˜ëŠ” ê°’:
        (í‰ê· , í•˜í•œ, ìƒí•œêµ¬ê°„)ìœ¼ë¡œ ì´ë£¨ì–´ì§„ tuple
    """
  
    data = np.array(data)
    mean = np.mean(data)
    n = len(data)
    s = data.std(ddof = 1)
    # stderr = stats.sem(data)
    # print(stderr)
    stderr = s / np.sqrt(n)
    print(stderr)
    
    t = stats.t.ppf( (1+confidence) / 2.0 , n-1)
    margin_of_error = t*stderr
    interval = stderr * stats.t.ppf((1 + confidence) / 2.0, n - 1)
    return (mean, mean - interval, mean + interval)


confidence_interval(coinflips)
```
