````python
# ì¼€ë¼ìŠ¤ ì™¸ì˜ í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤.
import numpy as np
import pandas as pd
import tensorflow as tf

# ë”¥ëŸ¬ë‹ì„ êµ¬ë™í•˜ëŠ” ë° í•„ìš”í•œ ì¼€ë¼ìŠ¤ í•¨ìˆ˜(ë©”ì„œë“œ)ë¥¼ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤.
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import tensorflow.keras.layers as Layer
````

2. **ì¬í˜„ì„ ìœ„í•´ì„œ ì‹œë“œë¥¼ ê³ ì •í•©ë‹ˆë‹¤. Numpy ë¿ë§Œ ì•„ë‹ˆë¼ Tensorflow ì˜ ì‹œë“œë„ ê³ ì •í•´ì£¼ì–´ì•¼ í•©ë‹ˆë‹¤.**

````python
# ì‹¤í–‰í•  ë•Œë§ˆë‹¤ ê°™ì€ ê²°ê³¼ë¥¼ ì¶œë ¥í•˜ê¸° ëœë¤í•¨ìˆ˜ë¥¼ ê³ ì •í•˜ëŠ” ë¶€ë¶„ì…ë‹ˆë‹¤.
# ëœë¤í•¨ìˆ˜ì˜ Seedë¥¼ ê³ ì •í•˜ê²Œ ë˜ë©´ ëœë¤í•¨ìˆ˜ê°€ í•­ìƒ ì¼ì •í•˜ê²Œ ë‚˜ì˜µë‹ˆë‹¤. 
np.random.seed(3)
tf.random.set_seed(3)
````

3. **ë°ì´í„°ë¥¼ ë¶ˆëŸ¬ì˜¨ í›„ Featureì™€ Label ì„ ë‚˜ëˆ„ì–´ ì¤ë‹ˆë‹¤.**

````python
# íì•” ìˆ˜ìˆ  í™˜ìì˜ íŠ¹ì •ê¸°ê°„ ìƒì¡´ ë°ì´í„°
# ì†ì„±(ì •ë³´)ì€ ì¢…ì–‘ì˜ ìœ í˜•, íí™œëŸ‰, í˜¸í¡ê³¤ë€ ì—¬ë¶€, ê¸°ì¹¨, í¡ì—°, ì²œì‹ì—¬ë¶€ ë“±ì˜ 17ê°€ì§€ í™˜ì ìƒíƒœ. ìˆ˜ìˆ  í›„ ìƒì¡´(1), ì‚¬ë§(0) 
my_data = "https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/everydeep/ThoraricSurgery.csv"
````

````python
# ë¶ˆëŸ¬ì˜¨ ë°ì´í„°ë¥¼ ì ìš©í•©ë‹ˆë‹¤.
# pandasì™¸ì—ë„ ì½ì„ ìˆ˜ ìˆëŠ” ë°©ë²•ì´ ìˆìŠµë‹ˆë‹¤. í¸í•˜ì‹  ë°©ë²•ì„ ì‚¬ìš©í•˜ì‹œë©´ ë©ë‹ˆë‹¤.
Data_set = np.loadtxt(my_data, delimiter=",") 
````

````python
# í™˜ìì˜ ê¸°ë¡ê³¼ ìˆ˜ìˆ  ê²°ê³¼ë¥¼ Xì™€ Yë¡œ êµ¬ë¶„í•˜ì—¬ ì €ì¥í•©ë‹ˆë‹¤.
X = Data_set[:,0:17]
Y = Data_set[:,17]
````

4. **ì‹ ê²½ë§ì„ êµ¬ì¶•í•©ë‹ˆë‹¤.**

> â—ï¸ ***ì´ë²ˆì—ë„ ì½”ë“œì—ì„œ ì¶œë ¥ì¸µì˜ ë…¸ë“œ ìˆ˜ëŠ” ëª‡ ê°œì¸ì§€, ì¶œë ¥ì¸µì˜ í™œì„±í™” í•¨ìˆ˜ëŠ” ë¬´ì—‡ì¸ì§€, ì†ì‹¤ í•¨ìˆ˜ëŠ” ì–´ë–»ê²Œ ì§€ì •í•˜ì˜€ëŠ”ì§€ì— ì£¼ëª©í•´ë´…ì‹œë‹¤.<br/>
ê·¸ë¦¬ê³  ì™œ ê·¸ë ‡ê²Œ ì§€ì •í•˜ì—¬ ì£¼ì—ˆëŠ”ì§€ ìƒê°í•´ë³´ê³  í† ë¡ í•´ë´…ì‹œë‹¤.***

````python
# ë”¥ëŸ¬ë‹ êµ¬ì¡°ë¥¼ ê²°ì •í•©ë‹ˆë‹¤(ëª¨ë¸ì„ ì„¤ì •í•˜ê³  ì‹¤í–‰í•˜ëŠ” ë¶€ë¶„ì…ë‹ˆë‹¤).
model = Sequential([
    Dense(30, activation='relu'),
    Layer.Dropout(0.5),
    Dense(30, activation='relu'),
    Dense(1, activation='sigmoid') # ë¶„ë¥˜í•  ë°©ë²•ì— ë”°ë¼ ê°œìˆ˜ë¥¼ ì¡°ì •í•´ì•¼ í•©ë‹ˆë‹¤. 
])
# ë”¥ëŸ¬ë‹ì„ ì‹¤í–‰í•©ë‹ˆë‹¤.
model.compile(loss='mean_squared_logarithmic_error', optimizer='adam', metrics=['accuracy']) # mean_squared_error # binary_crossentropy # mean_absolute_error # poisson
history = model.fit(X, Y, epochs=30, batch_size=30)
````

### ì†ì‹¤ í•¨ìˆ˜(Cost function) í˜¹ì€ ì˜¤ì°¨ í•¨ìˆ˜(Error function)

ë¬¸ì œì˜ ì¢…ë¥˜ì— ë”°ë¼ ì†ì‹¤ í•¨ìˆ˜ë¥¼ ì˜ ì„ íƒí•´ì£¼ì–´ì•¼ í•©ë‹ˆë‹¤.

í‰ê·  ì œê³±ì„ ê¸°ë°˜ìœ¼ë¡œ í•˜ëŠ” ì†ì‹¤ í•¨ìˆ˜ê°€ ìˆê³  ì—”íŠ¸ë¡œí”¼ë¥¼ ê¸°ë°˜ìœ¼ë¡œ í•˜ëŠ” ì†ì‹¤ í•¨ìˆ˜ê°€ ìˆìŠµë‹ˆë‹¤.

- **í‰ê·  ì œê³±ì„ ê¸°ë°˜ìœ¼ë¡œ í•˜ëŠ” ì†ì‹¤ í•¨ìˆ˜**

    1. MSE(Mean Squared Error)
$$
    \frac{1}{n}\sum_{i=1}^{n}(y_{i} - \hat{y_{i}})^{2}
$$

    2. RMSE(Root Mean Squared Error)
$$
    \sqrt{\frac{1}{n}\sum_{i=1}^{n}(y_{i} - \hat{y_{i}})^{2}}
$$

    3. MAE(Mean Absolute Error)
$$
    \frac{1}{n}\sum_{i=1}^{n}\left \vert y_{i} - \hat{y_{i}} \right \vert
$$

    4. ì´ ì™¸ì—ë„ R-Squared ë“±ì˜ ì˜¤ì°¨ í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤.

- **ì—”íŠ¸ë¡œí”¼ë¥¼ ê¸°ë°˜ìœ¼ë¡œ í•˜ëŠ” ì†ì‹¤ í•¨ìˆ˜**

    1. BCE(Binary Crossentropy)
$$
    -\sum_{c=1}^{C} q(y_c) log(q(y_c)), \quad q(y_c) \in (1, -1)
$$

    2. CCE(Categorical Crossentropy)
$$
    -\sum_{c=1}^{C} q(y_c)log(q(y_c)) 
$$

### Keras ë¡œ êµ¬ì¶•í•œ ì‹ ê²½ë§ì— ì¡°ê¸° ì¢…ë£Œ(Early Stopping) ì ìš©í•˜ê¸°

Fashion Mnist ë°ì´í„°ì…‹ì„ í•™ìŠµí•  ì‹ ê²½ë§ì— ì¡°ê¸° ì¢…ë£Œ(Early Stopping)ë¥¼ ì ìš©í•˜ì—¬ ë³´ê² ìŠµë‹ˆë‹¤.<br/>
ëŒ€ë¶€ë¶„ì˜ ì½”ë“œëŠ” ì´ì „ ê°•ì˜ì—ì„œ í–ˆë˜ ì˜ˆì œì™€ ë™ì¼í•©ë‹ˆë‹¤. Early Stopping ì´ ì ìš©ëœ 

1. **ë°ì´í„°ì…‹ì„ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤.**

````python
from tensorflow.keras.datasets import fashion_mnist

# ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
print(X_train.shape, X_test.shape)

# ë°ì´í„°ë¥¼ ì •ê·œí™” í•©ë‹ˆë‹¤
X_train = X_train / 255.
X_test = X_test /255.

# í´ë˜ìŠ¤ë¥¼ í™•ì¸í•©ë‹ˆë‹¤.
np.unique(y_train)
````

2. **í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ë¶ˆëŸ¬ì˜¨ í›„ ëª¨ë¸ì„ êµ¬ì¶•í•©ë‹ˆë‹¤.**

````python
# ê¸°ë³¸ì ì¸ ì‹ ê²½ë§ì„ ë§Œë“œëŠ” ì½”ë“œ
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
import keras, os

# ëª¨ë¸ êµ¬ì„±ì„ í™•ì¸í•©ë‹ˆë‹¤.
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(10, activation='softmax')
])
# ì—…ë°ì´íŠ¸ ë°©ì‹ì„ ì„¤ì •í•©ë‹ˆë‹¤.
model.compile(optimizer='adam'
             , loss='sparse_categorical_crossentropy'
             , metrics=['accuracy'])
model.summary()
# ì´ 7850 parameters (10 bias)
````

3. **ì¡°ê¸° ì¢…ë£Œ(Early Stopping)ë¥¼ ìœ„í•œ ì½œë°± í•¨ìˆ˜ë¥¼ ì§€ì •í•˜ê³  ëª¨ë¸ì„ í•™ìŠµí•©ë‹ˆë‹¤.**



````python
# ëª¨ë¸ í•™ìŠµì„ ìœ„í•œ ì½”ë“œ

# ë³€ìˆ˜ ì„¤ì •ì„ ë”°ë¡œ í•˜ëŠ” ë°©ë²•ì„ ì ìš©í•˜ê¸° ìœ„í•œ ì½”ë“œì…ë‹ˆë‹¤. 
batch_size = 30
epochs_max = 1

# í•™ìŠµì‹œí‚¨ ë°ì´í„°ë¥¼ ì €ì¥ì‹œí‚¤ê¸° ìœ„í•œ ì½”ë“œì…ë‹ˆë‹¤. 
checkpoint_filepath = "FMbest.hdf5"

# overfittingì„ ë°©ì§€í•˜ê¸° ìœ„í•´ì„œ í•™ìŠµ ì¤‘ early stopì„ ìˆ˜í–‰í•˜ê¸° ìœ„í•œ ì½”ë“œì…ë‹ˆë‹¤.
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', min_delta=0, patience=10, verbose=1)

# Validation Setì„ ê¸°ì¤€ìœ¼ë¡œ ê°€ì¥ ìµœì ì˜ ëª¨ë¸ì„ ì°¾ê¸° ìœ„í•œ ì½”ë“œì…ë‹ˆë‹¤.
save_best = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_filepath, monitor='val_loss', verbose=1, save_best_only=True,
    save_weights_only=True, mode='auto', save_freq='epoch', options=None)

# ëª¨ë¸ í•™ìŠµ ì½”ë“œ + early stop + Best model
model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs_max, verbose=1, 
          validation_data=(X_test,y_test), 
          callbacks=[early_stop, save_best])
````

4. **í•™ìŠµí•œ ëª¨ë¸ì„ í‰ê°€(Evaluation)í•©ë‹ˆë‹¤.**

````python
# í•™ìŠµëœ ëª¨ë¸ì„ ì´ìš©í•˜ì—¬ í…ŒìŠ¤íŠ¸í•˜ëŠ” ì½”ë“œ

model.predict(X_test[0:1])
test_loss, test_acc = model.evaluate(X_test,  y_test, verbose=2)
````

## ì¼€ë¼ìŠ¤(Keras)ì—ì„œ Regularization ì ìš©í•˜ê¸°

### ê°€ì¤‘ì¹˜ ê°ì†Œ(Weight Decay)

- ë‹¤ìŒê³¼ ê°™ì€ ì½”ë“œë¡œ ì´ë£¨ì–´ì§„ ê°€ì¤‘ì¹˜ ê°ì†Œë¥¼ ì ìš©í•˜ì—¬ ë³´ê² ìŠµë‹ˆë‹¤.<br/>
```
Dense(64, input_dim=64,
            kernel_regularizer=regularizers.l2(0.01),
            activity_regularizer=regularizers.l1(0.01))
```

- **L1 Regularization**

$$
    L(\theta_w) = \frac{1}{2} \sum_i (output_i - target_i)^2 + \color{blue}{\lambda} \cdot \color{red}{\vert \theta_w \vert}
$$ 

- **L2 Regularization**

$$
    L(\theta_w) = \frac{1}{2} \sum_i (output_i - target_i)^2 + \color{blue}{\lambda} \cdot \color{red}{\Vert \theta_w \Vert_2}
$$ 

ë¨¼ì € ê°€ì¤‘ì¹˜ ê°ì†Œ(Weight decay)ë¥¼ ì¼€ë¼ìŠ¤ì—ì„œ ìˆ˜í–‰í•˜ì—¬ ë³´ê² ìŠµë‹ˆë‹¤.<br/>
ê°€ì¤‘ì¹˜ ê°ì†Œ ë°©ì‹ì—ì„œëŠ” ìœ„ì™€ ê°™ì´ ì†ì‹¤ í•¨ìˆ˜ì— ê°€ì¤‘ì¹˜ì™€ ê´€ë ¨ëœ í•­ì„ ì¶”ê°€í•˜ì—¬ ì†ì‹¤ í•¨ìˆ˜ë¥¼ ìµœì í™” í•˜ëŠ” ê³¼ì •ì—ì„œ ê°€ì¤‘ì¹˜ê°€ ë„ˆë¬´ ì»¤ì§€ì§€ ì•Šë„ë¡ ì¡°ì •í•©ë‹ˆë‹¤.

ì•„ë˜ ê·¸ë¦¼ì„ ë³´ë©´ ê³¼ì í•©(Over-fitting)ì„ ë‚˜íƒ€ë‚¼ ë•Œ êµ¬ë¶ˆêµ¬ë¶ˆí•œ ê²½ê³„ë©´ì´ í˜•ì„±ë˜ëŠ” ê²ƒì„ í™•ì¸í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.<br/>
ì´ë ‡ê²Œ êµ½ì´ì¹˜ëŠ” ê·¸ë˜í”„ê°€ í˜•ì„±ë  ë•Œì˜ ê°€ì¤‘ì¹˜ëŠ” ì»¤ì§€ê¸° ë§ˆë ¨ì¸ë°ìš”.<br/>
ê°€ì¤‘ì¹˜ ê°ì†ŒëŠ” ì• ì´ˆë¶€í„° ê°€ì¤‘ì¹˜ê°€ í° ê°’ì„ ê°–ì§€ ëª»í•˜ê²Œ í•˜ì—¬ ê³¼ì í•©ì´ ë‚˜íƒ€ë‚˜ì§€ ì•Šë„ë¡ ë§Œë“­ë‹ˆë‹¤.

ìœ„ì—ì„œ ì œì‹œëœ ì†ì‹¤ í•¨ìˆ˜ ì‹ì„ ë‹¤ì‹œ ë³´ë„ë¡ í•˜ê² ìŠµë‹ˆë‹¤.<br/>
Regularization ì´ ì ìš©ë˜ì§€ ì•Šì€ ì†ì‹¤ í•¨ìˆ˜ëŠ” ì¶œë ¥ê°’(output)ê³¼ íƒ€ê²Ÿ(Target)ì˜ ì°¨ì´(ì˜ ì œê³±)ë§Œì„ ì†ì‹¤ë¡œ ê³„ì‚°í•©ë‹ˆë‹¤.<br/>
ê°€ì¤‘ì¹˜ ê°ì†Œë¥¼ ì ìš©í•˜ë©´ **ì†ì‹¤ í•¨ìˆ˜ì— ê°€ì¤‘ì¹˜$(\theta)$ë¡œ êµ¬ì„±ëœ í•­ì´ ì¶”ê°€**ë©ë‹ˆë‹¤.<br/>
ì´ë ‡ê²Œ ë˜ë©´ ì†ì‹¤ í•¨ìˆ˜ë¥¼ ìµœì í™” í•˜ëŠ” ê³¼ì •ì—ì„œ **ê°€ì¤‘ì¹˜$(\theta)$ì˜ í¬ê¸°ê°€ ì»¤ì§€ëŠ” ê²ƒì„ ë§‰ê¸° ë•Œë¬¸ì— ê³¼ì í•©ì„ ë°©ì§€**í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

1. **(ë°ì´í„°ëŠ” ì´ì „ì˜ ê²ƒì„ ê·¸ëŒ€ë¡œ ì‚¬ìš©í•˜ë¯€ë¡œ) ê°€ì¤‘ì¹˜ ê°ì†Œê°€ ì ìš©ëœ ì‹ ê²½ë§ ëª¨ë¸ì„ êµ¬ì¶•í•©ë‹ˆë‹¤.**

````python
# ê°€ì¤‘ì¹˜ ê°ì†Œ(Weight Decay)ë¥¼ ë°˜ì˜í•œ ì˜ˆì‹œ ì½”ë“œ
from tensorflow.keras import regularizers

# ëª¨ë¸ êµ¬ì„±ì„ í™•ì¸í•©ë‹ˆë‹¤.
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(64, 
            kernel_regularizer=regularizers.l2(0.01),    # L2 Regularization ì ìš©
            activity_regularizer=regularizers.l1(0.01)), # L1 Regularization ì ìš© (kernel, activity ì˜ ì°¨ì´ì— ëŒ€í•´ì„œëŠ” ì¼ë‹¨ ë„˜ì–´ê°‘ì‹œë‹¤.)
    Dense(10, activation='softmax')
])
````

````python
# ì—…ë°ì´íŠ¸ ë°©ì‹ì„ ì„¤ì •í•©ë‹ˆë‹¤.
model.compile(optimizer='adam'
             , loss='sparse_categorical_crossentropy'
             , metrics=['accuracy'])
````

### ê°€ì¤‘ì¹˜ ì œí•œ(Weight Constraints)

> â—ï¸ ***ì¼€ë¼ìŠ¤ì—ì„œëŠ” ì—¬ëŸ¬ ê°€ì§€ ê°€ì¤‘ì¹˜ ì œí•œ ë°©ë²•ì„ ì œê³µí•˜ê³  ìˆìŠµë‹ˆë‹¤. ë‚˜ì¤‘ì— [ë§í¬](https://keras.io/api/layers/constraints/)ë¥¼ ì°¸ê³ í•´ë³´ë„ë¡ í•©ì‹œë‹¤.<br/>
(ë‹¹ì—°íˆ ì§€ê¸ˆ ëª¨ë“  ê²ƒì„ ì•Œì•„ì•¼ í•  í•„ìš”ëŠ” ì—†ìŠµë‹ˆë‹¤.)***

1. **ê¸°ì¡´ ì‹ ê²½ë§ì— ê°€ì¤‘ì¹˜ ì œí•œì´ ì¶”ê°€ë¡œ ì ìš©ëœ ì‹ ê²½ë§ ëª¨ë¸ì„ êµ¬ì¶•í•©ë‹ˆë‹¤.**

````python
# ê°€ì¤‘ì¹˜ ì œí•œ(Weight Constraints)ë¥¼ ë°˜ì˜í•œ ì˜ˆì‹œ ì½”ë“œ
from tensorflow.keras.constraints import MaxNorm

# ëª¨ë¸ êµ¬ì„±ì„ í™•ì¸í•©ë‹ˆë‹¤.
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(64, 
            kernel_regularizer=regularizers.l2(0.01),
            activity_regularizer=regularizers.l1(0.01),
            kernel_constraint=MaxNorm(2.)),             # ê°€ì¤‘ì¹˜ ì œí•œ(Weight Constraints)ì„ ì¶”ê°€ë¡œ ì ìš©
    Dense(10, activation='softmax')
])
````
